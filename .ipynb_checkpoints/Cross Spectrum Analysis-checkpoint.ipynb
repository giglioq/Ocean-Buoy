```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
t = np.linspace(0,50,1000)
```

```python
x = np.sin(t)
```

```python
y = np.sin(3*t)
```

```python
plt.plot(x)
plt.plot(y)
```

```python
Fx = fftpack.fft(x) 
Fy = fftpack.fft(y)
```

```python
plt.plot(Fx.real)
plt.plot(Fy.real)
```

```python
Cxy = Fx.real*Fy.real + Fx.imag*Fy.imag
```

```python
plt.plot(Cxy)
```

```python
Qxy = Fx.real*Fy.imag - Fx.imag*Fy.real
```

```python
plt.plot(Qxy)
```

```python
enumerate(Cxy)
```

```python
cross = [np.complex(i, j) for index, [i,j] in zip(enumerate(Cxy),enumerate(Qxy))]
```

```python
plt.plot(cross)
```
